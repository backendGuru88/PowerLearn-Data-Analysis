```python

#Plotting for PetalLengthCm lineChart

plt.plot(df['PetalLengthCm'],df['PetalLengthCm'], color='red')
plt.xlabel('PetalLengthCm')
plt.ylabel('Species')


plt.title('PetalLengthCm')
plt.show()
```

```python
# Plotting for barchart

df.groupby('Species')['PetalLengthCm'].mean().plot(kind = 'bar', color = 'red')
plt.xlabel('Species')
plt.ylabel('Average PetalLengthCm')
plt.title('Average petal length per species')
plt.show()
```

```python
# Plotting for Histogram

df['PetalLengthCm'].plot(kind='hist', color = 'red')
plt.xlabel('PetalLengthCm')
plt.title("Distribution")
plt.show()
```

```python
df.plot(kind = 'scatter', x = 'Species', y = 'PetalLengthCm')
plt.xlabel('Species')
plt.ylabel('PetalLengthCm')
plt.show()
```
